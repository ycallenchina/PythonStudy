```python
# 标准答案

class Shape:

    def perimeter(self):

        pass



    def area(self):

        pass



class Square(Shape):

    def __init__(self, a):

        self.a = a



    def perimeter(self):

        print('正方形周长是：', self.a * 4)



    def area(self):

        print('正方形面积是：', self.a ** 2)



class Rect(Shape):

    def __init__(self, a, b):

        self.a = a

        self.b = b



    def perimeter(self):

        print('长方形周长是：', (self.a + self.b) * 2)



    def area(self):

        print('长方形面积是：', self.a * self.b)



class Cube(Shape):

    def __init__(self, a):

        self.a = a



    def area(self):

        print('立方体面积是：', self.a ** 2 * 6)



    def volume(self):

        print('立方体体积是：', self.a ** 3)



squa = Square(5)

rect = Rect(5, 2)

cube = Cube(5)

squa.perimeter()

squa.area()

rect.perimeter()

rect.area()

cube.area()

cube.volume()
```

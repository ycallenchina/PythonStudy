```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 10)
sh=pd.read_csv(r'C:\Users\YcAllenEffy\Desktop\data30.csv')
sh
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
sns.set_style('dark',{'font.sans-serif':['simhei','Arial']})#初始化表格风格,字体设置
```

```python
sns.scatterplot(x='浪高上限',y='风速上限',data=sh)
```

```python
sns.scatterplot(x='浪高上限',y='风速上限',data=sh,hue='浪向')
```

```python
co=sns.color_palette('Blues',11)
sns.set_palette(co)
```

```python
sns.set_style('ticks',{'font.sans-serif':['simhei','Arial']})#初始化表格风格,字体设置

sns.barplot(x='风向',y='浪高上限',data=sh)
sns.despine(left=True)
```

```python
fig,ax=plt.subplots(2,2,figsize=(15,9))

sns.barplot(x='航线',y='浪高上限',data=sh,ax=ax[0][0])
sns.barplot(x='航线',y='风速上限',data=sh,ax=ax[0][1])
sns.barplot(x='浪向',y='浪高上限',data=sh,ax=ax[1][0])
sns.barplot(x='浪向',y='风速上限',data=sh,ax=ax[1][1])
sns.despine(offset=20)
```

```python
# 标准答案
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid', {'font.sans-serif':['simhei', 'Arial']})
# df = pd.read_excel(r'.\气象观测.xlsx')

df_t = df[df['站点'].isin(['G3523', 'G3540', 'G3531'])]

fig, axs = plt.subplots(1, 2, figsize=(20, 7))
sns.boxplot(data=df_t, x='站点', y='气温', ax=axs[0])
sns.swarmplot(data=df_t, x='站点', y='气温', ax=axs[1])
sns.boxenplot(data=df_t, x='站点', y='气温', ax=axs[1], 
             outlier_prop=0.1, k_depth='proportion');
```

###  下载并读取气象观测数据，完成以下需求：

    1. 创建箱型图，展示5月份的风速观测数据；

    2. 创建分类散点图，展示问题1中的风速观测数据；

    3. 在分类散点图绘图区域上创建增强箱型图，展示风速观测数据，设置异常点的范围在2%。   

```python
df['时间']=pd.to_datetime(df['时间'])
df['月份']=df['时间'].dt.month
df
```

```python
df_5=df[df['月份']==5]
sns.boxplot(y='风速',data=df_5)
```

```python
g=sns.stripplot(y='风速',data=df_5,color='red')
g=sns.boxenplot(y='风速',data=df_5,outlier_prop=0.02,k_depth='proportion')
g.figure.set_size_inches(5, 15) #图形大小设置
```

```python
# 标准答案
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid', {'font.sans-serif':['simhei', 'Arial']})
# df = pd.read_excel(r'.\气象观测.xlsx')

df['时间'] = pd.to_datetime(df['时间'])
df['月'] = df['时间'].dt.month
df_t = df[df['月']==5]

fig, axs = plt.subplots(2, 1, figsize=(20, 10))
sns.boxplot(data=df_t, x='风速', ax=axs[0])
sns.stripplot(data=df_t, x='风速', ax=axs[1], palette='Reds')
sns.boxenplot(data=df_t, x='风速', ax=axs[1], 
             outlier_prop=0.02, k_depth='proportion');
```

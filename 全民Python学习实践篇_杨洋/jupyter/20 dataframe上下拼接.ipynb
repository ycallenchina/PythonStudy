### Series 的上下拼接

  sr1.append(sr2)  规则:保留sr1和sr2全部索引,即使重复.

  sr1.append(sr2,ignore_index=True) 建立新的不重复索引



### Dataframe 的上下拼接 

  append 规则:索引即使相同也保留,列名如果相同就共用.

  df11.append(df2) 


---
  concat 规则:默认与append一致

  pandas.concat([df1,df2]) 效果与append一致,其中axis=0为默认值

  pandas.concat([df1,df2],ignore_index=True) 建立新的不重复索引

  pandas.concat([df1,df2],axis=1) axis=1时:列名即使相同也保留,但是行名有相同就共用

  pandas.concat([df1,df2],axis=1,join='inner') join='inner'为内连接:只显示df1和df2共用的列名, 

  默认为join='outer' 全连接,显示所有.

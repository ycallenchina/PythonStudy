### 下载并读取学校数据，完成以下需求：

    1. 创建箱型图和散点图，在同一个图表中展示各门课程的平均成绩分布；

    2. 创建箱型图和散点图，在同一个图表中展示各名教师负责课程的最低成绩分布。

```python
df_mean=df_st.groupby('课程')[['分数']].mean()
df_mean['课程2']=df_mean.index

sns.boxplot(y=df_mean['分数'])
sns.swarmplot(y='分数',data=df_mean,size=10,palette='prism')
```

```python
df_min=df_st.groupby('教师编号')[['分数']].min()
sns.boxplot(y=df_min['分数'])
sns.swarmplot(y='分数',data=df_min,size=10,palette='prism')
```

```python
# 标准答案
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('whitegrid', {'font.sans-serif':['simhei', 'Arial']})

# df = pd.read_excel(r'.\school.xlsx', sheet_name=[0,1,2,3])
df_stu = df[0]
df_course = df[1]
df_teacher = df[2]
df_score = df[3]
df_t = pd.merge(pd.merge(df_score, df_stu, on='学生编号'), df_course, on='课程编号')
df_t = pd.merge(df_t, df_teacher, on='教师编号')

fig, axs = plt.subplots(1, 2, figsize=(16, 7))

df_ss = df_t.groupby('课程')[['分数']].mean()
sns.boxplot(data=df_ss, y='分数', ax=axs[0])
sns.swarmplot(data=df_ss, y='分数', ax=axs[0], palette='prism', size=10)

df_ts = df_t.groupby('教师姓名')[['分数']].min()
sns.boxplot(data=df_ts, y='分数', ax=axs[1])
sns.swarmplot(data=df_ts, y='分数', ax=axs[1], palette='prism', size=10);
```

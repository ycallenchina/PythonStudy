```python
# 标准答案
import pandas as pd
from pyecharts.charts import Map
from pyecharts import options as opts

dfp = sh.copy()

iopt = opts.InitOpts(width='1920px', height='1080px')
pop_map = Map(iopt)

lopt = opts.LabelOpts(is_show=True, formatter='{b}\n{c} 亿元')
pop_map.add('2021年第一季度生产总值', list(zip(dfp['地区'], dfp['2021年第一季度'])), maptype='china', label_opts=lopt)

pop_map.render('2021年第一季度生产总值.html')
```

###     使用 常住人口数据，按以下要求完成练习：

    1. 创建地图并设置大小；

    2. 配置热力地图设置最大值为人口最大值，最小值为人口最小值；

    3. 绘制广州各区的人口数据，格式化标签显示地区和人口，单位万人。

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 20)
sh=pd.read_csv(r'C:\Users\YcAllenEffy\Desktop\常住人口数据.csv')
df=sh.copy()
```

```python
# 初始化
iopt=opts.InitOpts(width='1080px',height='500px')
lopt=opts.LabelOpts(is_show=True,formatter='{b}\n{c}万人')

# 热力图
vopt=opts.VisualMapOpts(min_=df['人口'].min(),max_=df['人口'].max())
```

```python
p_map=Map(iopt)
p_map.add('常住人口',list(zip(df['地区'],df['人口'])),maptype='广州',label_opts=lopt)
p_map.set_global_opts(visualmap_opts=vopt)
```

```python
p_map.render_notebook()
```

```python
# 标准答案
import pandas as pd
from pyecharts.charts import Map
from pyecharts import options as opts

dfp = sh.copy()
iopt = opts.InitOpts(width='1920px', height='1080px')
pop_map = Map(iopt)

vopt = opts.VisualMapOpts(min_=dfp['人口'].min(), max_=dfp['人口'].max())
pop_map.set_global_opts( visualmap_opts=vopt )

lopt = opts.LabelOpts(is_show=True, formatter='{b}\n{c} 万人')
pop_map.add('常住人口数据', list(zip(dfp['地区'], dfp['人口'])), maptype='广州',
            label_opts=lopt)

pop_map.render_notebook()
```

###     使用 国际卖场数据，按以下要求完成练习：

    1. 创建世界热力地图，绘制各国家地区的总额；

    2. 通过修改数据来显示中国地区的总额；

    3. 观察绘制的地图中是否只有一个国家的颜色是在热力图图例中属于偏上的颜色，思考如何处理这样差距过大的数据，能够让其他国家的总额能够展示出热力图图例偏上的颜色，比如橙色到红色之间。

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 20)
sh=pd.read_csv(r'C:\Users\YcAllenEffy\Desktop\国际卖场数据new.csv')
df=sh.copy()
```

```python
df['国家地区']=df['国家地区'].str.replace('China Mainland','China')
dfs = df.groupby('国家地区').sum()
```

```python
iopt = opts.InitOpts(width='1800px', height='1000px')
p_map = Map(iopt)
lopt=opts.LabelOpts(is_show=False)
vopt = opts.VisualMapOpts(min_=dfs['总额'].min(), max_=dfs['总额'].max())
p_map.add('世界卖场',list(zip(dfs.index,dfs['总额'])),maptype='world',label_opts=lopt)
p_map.set_global_opts(visualmap_opts=vopt)
p_map.render_notebook()
```

```python
# 标准答案
import pandas as pd
from pyecharts.charts import Map
from pyecharts import options as opts

df =sh.copy()
df['国家地区'] = df['国家地区'].str.replace('China Mainland', 'China')

dfs = df.groupby('国家地区').sum()

# 地区1
pop_map = Map()
vopt = opts.VisualMapOpts(min_=dfs['总额'].min(), max_=dfs['总额'].max())
pop_map.set_global_opts( visualmap_opts=vopt )
lopt = opts.LabelOpts(is_show=False)
pop_map.add('国际卖场', list(zip(dfs.index, dfs['总额'])), maptype='world',
            label_opts=lopt)
pop_map.render_notebook()

```

```python
# 地图2
dfs.sort_values('总额', ascending=False)
dfs.loc['Australia', '总额'] = dfs.loc['Australia', '总额'] * 2.2
pop_map2 = Map()
pop_map2.set_global_opts( visualmap_opts=vopt )
pop_map2.add('国际卖场', list(zip(dfs.index, dfs['总额'])), maptype='world',
            label_opts=lopt)
pop_map2.render_notebook()
```

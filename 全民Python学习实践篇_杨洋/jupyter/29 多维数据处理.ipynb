### 多维数据提取

  #### 最外层索引,列名直接读取

    df['最外层列名']

    df.loc['最外层索引']

    df.loc['最外层索引','最外层列名']

  #### 内层索引,内层列名需要逐级读取

    df[('最外层列名','第二层列名')]

    df.loc[('最外层索引','第二层索引')]

    **df.loc[(索引级),(列名级)]**

      df.loc[('最外层索引','第二层索引'),('最外层列名','第二层列名')]

  #### 层级互换

    df.swaplevel()

      默认:axis=0 为索引的层级交换

           axis=1为列的层级交换
       注意: 如果层级超过3层,那么只会互换最里层2层级.

    **读取 最外层索引,所有的第二层列名的数据方法**

      df=df.swaplevel(axis=1)

      df.loc['最外层索引','第二层列名']

### 多维数据统计

  **df.统计函数()**  df.sum()

|统计函数()|参数axis=0|参数axis=1|参数<br>不写|
|-|-|-|-|
|参数level=0|计算 **行** 的层级为0级,的列总和,计算方向为向下.|计算列的层级为0级,的列总和,计算方向为向下.|axis不写时,默认=0|
|参数level=1|计算 **行** 的层级为1级,的列总和,计算方向为向下.|计算列的层级为1级,的列总和,计算方向为向下.|axis不写时,默认=0|
|参数<br>不写|axis=0 时,(默认),计算所有列的总和,计算方向向下.|axis=1 时,计算所有行的总和,计算方向向左.|axis和level都不写时,默认axis=0效果|

### 行列转换

#### 列名转为索引

  df=df.stack() 默认将最内层列名转为最内层索引

  df=df.stack(level=0) 将第0层列名转为最内层索引

#### 索引转为列名

  df=df.unstack() 默认将最内层索引转为最内层列名

  df=df.stack(level=0) 将第0层索引转为最内层列名

###  使用 银行交易日志 数据，按以下要求完成练习：

    >=20万 -> '50'

    注意：有部分账户节点不属于发起账户，这一部分节点设置为 '3'。

```python
import pandas as pd
from pyecharts.charts import Graph

file_name = r'银行交易日志.xlsx'
dfn = sh.copy()

nodes = dfn['发起账户'].append( dfn['对方账户'] ).drop_duplicates()
nodes = [ {'name': i} for i in nodes ]

# 观察总账户余额
dfs = dfn.groupby('发起账户').sum()
dfs = dfs.sort_values('账户余额', ascending=False)

# 根据规则设置节点图形大小
for i in nodes:
    name = i.get('name')
    if name in list(dfs.index): # 如果节点姓名属于发起账户
        balance = dfs.loc[name]['账户余额']
        t = ''
        if balance >= 200000:
            t = '50'
        elif balance >= 100000:
            t = '30'
        elif balance >= 50000:
            t = '15'
        else:
            t = '5'
    else:
        t = '3'
    i['symbolSize'] = t

links = []
for i in range(len(dfn)):
        links.append( { 'source': dfn.iloc[i]['发起账户'], 'target': dfn.iloc[i]['对方账户'] } )

g = Graph()
g.add( '交易关系', nodes, links, layout='force', repulsion=200, edge_symbol=['', 'arrow'] )
g.render_notebook()
```

### 整体运算

  df*5
  支持 广播机制

#### 统计函数

  np.sum(df) 按列求和

  np.sum(df.values) 求总和

  np.sum(df,axis=1) 按行求和
  
  df.sum(axis=0) 按行求和(默认axis=0)
   
  df.mena(axis=0)
#### 对位运算

  规则:

  机制:

  df.add(df2) df对位相加df2

  df.add(df2,fill_value=0) df对位相加df2 缺失值补0处理

  其他df方法:sub 减,mul乘,div除,mod求模,pow乘方,floordiv整除

  df.radd 方法: 反向检查,可以处理例如 fill_value=0 时结果还是NaN

#### df与series的计算

  规则:df对series运算时,会把其看成一行,做列运算.

  df*series

  做行运算方法: 
      
     相乘:df.multiply(series,axis=0)

        其中axis也可以='index'代替0 意思按照索引行运算

         axis可以 ='colunms' 代替1 按照列运算

```python
import pandas as pd
import numpy as np
boolean=[True,False]
gender=["男","女"]
color=["white","black","yellow"]
df1=pd.DataFrame({
    "height":np.random.randint(150,190,10),
    "weight":np.random.randint(40,90,10),
    "smoker":[boolean[x] for x in np.random.randint(0,2,10)],
    "gender":[gender[x] for x in np.random.randint(0,2,10)],
    "age":np.random.randint(15,90,10),
    "color":[color[x] for x in np.random.randint(0,len(color),10) ]
    }
) 
df1
```

### 整体运算

```python
df2=df1[['height','weight']]
df2
```

```python
df2*2
```

### 统计函数 

```python
np.sum(df2)
```

```python
np.sum(df2.values)
```

```python
np.sum(df2,axis=1)
```

### 对位运算 

```python
df3=df1[['age','weight']] 
df3
```

```python
df2.add(df3) #只有索引与 列名对上的值才能相加
```

```python
df2.add(df3,fill_value=0) #空值补0
```

```python
df2.sub(df3,fill_value=0) #df2减df3空值补0
```

### df与series运算 

```python
sr1=df1['age']
sr1
```

```python
df2+sr1
```

```python
 df2.add(sr1,axis='index') #df2加sr1 按行相加
```

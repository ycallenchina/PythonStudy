```python
# 标准答案
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

ser1 = np.random.normal(0, 1.0, 1000)
ser2 = np.exp(ser1)

fig, axs = plt.subplots(1, 2, figsize=(20, 7))
sns.histplot(x=ser1, kde=True, ax=axs[0])
sns.histplot(x=ser2, kde=True, ax=axs[1])
```

```python
# 标准答案
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('whitegrid', {'font.sans-serif':['simhei', 'Arial']})

df = sh

# 1. 绘制整体的分布
sns.histplot(data=df, x='成绩', kde=True)
# 2. 拆分数据集
sns.histplot(data=df[ df['性别']=='男' ], x='成绩', kde=True)
sns.histplot(data=df[ df['性别']=='女' ], x='成绩', kde=True)
# 或者使用hue参数
sns.histplot(data=df, x='成绩', kde=True, hue='性别')
```

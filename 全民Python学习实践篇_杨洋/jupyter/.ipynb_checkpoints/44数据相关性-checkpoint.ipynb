###  读取卖场数据，按照以下要求完成练习：

    1. 筛选商品子类为Fasteners的数据，使用散点图绘制运费与总额的相关性。

    2. 使用pairplot方法绘制 '总额', '运费', '折扣', '营利', '数量', '运输等级' 等指标的相关性，按照运输等级设置图例。



```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 20)
sh=pd.read_csv(r'C:\Users\YcAllenEffy\Desktop\国际卖场.csv')
df=sh.copy()
df
```

```python
df=df[df['商品子类']=='Fasteners']
df
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
#设置画图 中文字体初始化
sns.set_style('whitegrid', {'font.sans-serif':['simhei', 'Arial']})
```

```python
sns.scatterplot(data=df,x='运费',y='总额')
```

```python
df2=df[['总额','运费','折扣','营利','数量','运输等级']]
sns.pairplot(data=df2,hue='运输等级')
```

```python
# 标准答案
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid', {'font.sans-serif':['simhei', 'Arial']})
plt.rcParams['axes.unicode_minus'] = False

df = sh.copy()
dfc = df[df['商品子类']=='Fasteners']
dfc = dfc[['总额', '运费', '折扣', '营利', '数量', '运输等级']]

sns.scatterplot(data=dfc, x='运费', y='总额')
sns.pairplot(data=dfc, hue='运输等级')
```

```python
# 标准答案
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid', {'font.sans-serif':['simhei', 'Arial']})
plt.rcParams['axes.unicode_minus'] = False

df = sh.copy()
dfc = df[df['商品子类']=='Fasteners']
dfc = dfc[['总额', '运费', '折扣', '营利', '数量', '运输等级']]
sns.relplot(data=dfc, x='运费', y='总额')
sns.relplot(data=dfc, x='运费', y='总额', kind='line', hue='运输等级')
```

###    继续使用练习2的数据集，按照以下要求完成练习：

    1. 使用jointplot方法绘制数量与总额的相关性；

    2. 将图表改为kde曲线图，设置阴影，颜色为蓝色。

```python
sns.jointplot(data=df2,x='数量',y='总额')
```

```python
sns.jointplot(data=df2,x='数量',y='总额',kind='kde',shade=True,color='blue')
```

```python
# 标准答案
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid', {'font.sans-serif':['simhei', 'Arial']})
plt.rcParams['axes.unicode_minus'] = False

df = sh.copy()
dfc = df[df['商品子类']=='Fasteners']
dfc = dfc[['总额', '运费', '折扣', '营利', '数量', '运输等级']]
sns.jointplot(data=dfc, x='数量', y='总额')
sns.jointplot(data=dfc, x='数量', y='总额', kind='kde', shade=True, color='b')
```

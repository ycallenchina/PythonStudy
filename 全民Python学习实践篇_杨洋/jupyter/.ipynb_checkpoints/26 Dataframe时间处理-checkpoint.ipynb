#### 时间查询

  注意:首先需要设置好时间所有,并给索引排好序.

    df.loc['1900/01/01']  只有日期,则会查询所有包含在该日期的时间

    df.loc['1900/01/01 15'] 查询1900年1月1日所有包含15时的记录

  切片

    df.loc['1900/01/01':'1900/01/01 15'] 格式不做约束,规则按照范围查询



#### 时间的提取

  df['时间列'].dt.day 提取时间列所有日

  df['时间列'].dt.month 提取时间列所有月份

  df['时间列'].dt.dayofweek 提取时间列所有星期几

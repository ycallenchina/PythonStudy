### 量比计算

#### 时间选择

```python
# 时间选择
df_open = pro.query('trade_cal', start_date='20210803', end_date='20211031')#读取股市开盘时间
date=df_open[df_open['is_open']==1]['cal_date']#选择开盘时间列

#初始化设置
df_date=pd.DataFrame(date) #转换位dateframe
df_date['cal_date']=pd.to_datetime(df_date['cal_date']) #转换位datetime格式
df_date.index=df_date['cal_date'] #设置为时间索引 方便选择

# 选择时间
df_date=df_date.loc['2021-09':'2021-11']

#输出对应时间格式
date_select=df_date['cal_date'].apply(lambda x:x.strftime('%Y%m%d'))
date_select.count()
```

#### 按照股票分类

```python
# 设置量比列
df=sh.copy()
df['量比']=0

# 读取不重复股
df_gb_stock=df.groupby('ts_code').sum()
df_gb_stock

df
```

#### 量比计算

```python
# 计算量化

# 读取每个股票的历史
# df_gb_stock为个股名录.
erro_count=0
for i in df_gb_stock.index:#循环所有股
    df_symbol=df[df['ts_code']==i]
    
    the_last=-1
    for k in range(36):#循环个股36个交易日时间
        try:
            self_amount=df_symbol.iloc[the_last,-2]#自身成交量
            front_5mount_mean=df_symbol.iloc[the_last-5:the_last,-2].mean()#前5条成交量均值
            
            #定位当前纪录的索引于df里
            record_index=df_symbol.iloc[[the_last]].index.tolist()[0]
            #修改此纪录的amount为量比计算结果
            df.loc[record_index,'量比']=round(self_amount/front_5mount_mean,2)
            
            #下一条纪录
            the_last+=-1
            
        except:
            print('erro',erro_count)
            erro_count+=1
            pass
```

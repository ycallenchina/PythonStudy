### 下载并读取证券数据，筛选证券代码、证券简称、网上发行日、发行量等4列数据，获取单日最高发行量中前10名股票的相关数据，按发行量从高到低显示。 

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', None)
sh3=pd.read_excel(r'C:\Users\YcAllenEffy\Desktop\网易新股.xlsx',header=1,usecols='B:I',index_col='序号')
sh3
```

```python
df_4col=sh3[['证券代码','证券简称','网上发行日','发行量(万股)']]
```

```python
df_4col.sort_values('网上发行日')
```

###  注意:df分组,如统计某一列值max,求单值的,再外加显示别的列名时候,并不一定会对应显示max单值对应的其他列名值.

```python
df_top10=df_4col.groupby('网上发行日')[['发行量(万股)']].max().sort_values('发行量(万股)',ascending=False).head(10)
```

```python
df_4col[df_4col['发行量(万股)'].isin(df_top10['发行量(万股)'])].sort_values('发行量(万股)',ascending=False)
```

### 下载并读取学生与分数数据，获取总分排在最高的前两名学生的学生编号、学生姓名、年龄、性别、总分。

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', None)
sh4=pd.read_excel(r'C:\Users\YcAllenEffy\Desktop\school.xlsx',sheet_name=[0,1,2,3])
sh4[3]
```

```python
df_score=sh4[3]
```

```python
df_sum=df_score.groupby('学生编号')[['分数']].sum()
```

```python
df_top2=pd.merge(sh4[0],df_sum,on='学生编号').sort_values('分数').tail(2)
```

```python
df_top2.rename(columns={'分数':'总分'}, inplace = True)
```

```python
df_top2
```

```python
# 标准答案
df_stu = sh4[0]
df_score =sh4[3]

df_top2 = df_score.groupby('学生编号')[['分数']].sum().sort_values('分数', ascending=False).head(2)
df_res = df_stu[df_stu['学生编号'].isin(df_top2.index)]
df_res.merge(df_top2, left_on='学生编号', right_index=True)
```

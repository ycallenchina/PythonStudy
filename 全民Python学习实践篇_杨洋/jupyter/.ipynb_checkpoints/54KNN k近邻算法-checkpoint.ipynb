### 读取 火焰纹章角色数据-1，使用KNN算法推断未知角色“凯尔”的职业。

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 20)
sh=pd.read_excel(r'C:\Users\YcAllenEffy\Desktop\fire_emblem_1.xlsx',engine='openpyxl',sheet_name=0)
df=sh.copy()
df
```

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 20)
sh2=pd.read_excel(r'C:\Users\YcAllenEffy\Desktop\fire_emblem_1.xlsx',engine='openpyxl',sheet_name=1)
df_target=sh2.copy()
df_target
```

```python
r=df_target.loc[0]
```

```python
# 算出每行数据的knn值
all_dst=df.apply(lambda x:np.sqrt(sum([ (r[i]-x[i])**2 for i in range(2,12) ])),axis=1)
```

```python
# 排序
df.loc[all_dst.sort_values().head(10).index,:]['职业']
```

```python
# 标准答案
import pandas as pd
import numpy as np

file = r'C:\Users\YcAllenEffy\Desktop\fire_emblem_1.xlsx'
df_src = pd.read_excel(file, sheet_name=0)
df_unknown = pd.read_excel(file, sheet_name=1)

r = df_unknown.iloc[0, :] # 未知点
x = df_src.iloc[0, :]     # 已知点

all_dst = df_src.apply( 
    lambda x: np.sqrt( sum( [(r[i]-x[i])**2 for i in range(2, 12)] ) ), axis=1 )

k = 5
res = df_src.loc[ all_dst.sort_values().head(k).index, : ]['职业'].mode()[0]
print('凯尔的职业：', res)
```

```python
# 标准答案
import pandas as pd
import numpy as np

file =r'C:\Users\YcAllenEffy\Desktop\fire_emblem_3.xlsx'
df_src = pd.read_excel(file, sheet_name=0)
df_unknown = pd.read_excel(file, sheet_name=1)

r = df_unknown.iloc[0, :] # 未知点
x = df_src.iloc[0, :]     # 已知点

all_dst = df_src.apply( 
    lambda x: np.sqrt( sum( [(r[i]-x[i])**2 for i in range(2, 11)] ) ), axis=1 )

k = 5
res = df_src.loc[ all_dst.sort_values().head(k).index, : ]['行动点'].mean()
print('芬里尔的行动点：', res)
```

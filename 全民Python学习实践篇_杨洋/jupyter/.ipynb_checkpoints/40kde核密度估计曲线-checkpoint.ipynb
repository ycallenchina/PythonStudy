```python
# 标准答案
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('whitegrid', {'font.sans-serif':['simhei', 'Arial']})

df = sh
df_t = df.groupby('站名')[['最低能见度']].mean()

fig, axs = plt.subplots(1, 2, figsize=(20, 7))
g1 = sns.violinplot( data=df_t, x='最低能见度', ax=axs[0])
g2 = sns.violinplot( data=df_t, y='最低能见度', inner='quartile', ax=axs[1])
_ = plt.setp(g1.collections, alpha=1)
_ = plt.setp(g2.collections, alpha=0.6)
```

###      下载并读取能见度数据，完成以下需求：

    1. 创建提琴图，垂直方向分别展示“东部华侨城”、“立新水库”、“妈湾港”三个站点的能见度分布情况，且只展示样本范围内的情况；

    2. 创建提琴图，垂直方向展示问题1中的三个站点的最低能见度分布情况，且只展示样本范围内的情况。

```python
df1=df[df['站名'].isin(['东部华侨城','立新水库','妈湾港']) ]
sns.violinplot(data=df1,x='站名',y='能见度',cut=0)
```

```python
sns.violinplot(data=df1,x='站名',y='最低能见度',cut=0)
```

```python
# 标准答案
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('whitegrid', {'font.sans-serif':['simhei', 'Arial']})

df = sh

df_t = df[df['站名'].isin(['东部华侨城', '立新水库', '妈湾港'])]

fig, axs = plt.subplots(1, 2, figsize=(20, 7))
g1 = sns.violinplot( data=df_t, x='站名', y='能见度', cut=0, ax=axs[0])
g2 = sns.violinplot( data=df_t, x='站名', y='最低能见度', cut=0, ax=axs[1])
_ = plt.setp(g1.collections, alpha=0.6)
_ = plt.setp(g2.collections, alpha=0.6)
```

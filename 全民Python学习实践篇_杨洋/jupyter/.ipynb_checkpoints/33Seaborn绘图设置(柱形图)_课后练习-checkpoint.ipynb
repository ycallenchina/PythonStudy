###    下载并读取国际卖场数据，完成以下需求：

    1. 创建三张柱形图，分别展示销售总额排名前5的国家的"办公"销售总额、"家具"销售总额、"科技"销售总额；

    2. 将三个柱形图绘制到同一行。 

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 20)
sh=pd.read_csv(r'C:\Users\YcAllenEffy\Desktop\国际卖场数据2011-2015.csv')
sh
```

```python
# 找出销售总额前5国家
df=sh
df_top5=df.groupby('国家地区')[['总额']].sum().sort_values('总额',ascending=False).head(5)
df_top5
```

```python
# 找出前5国家的 "办公"销售总额、"家具"销售总额、"科技"销售总额；
df['商品门类'].unique()
```

```python
# 前5国办公销售总额
df_off=df[df['商品门类']=='Office Supplies']
df_off5=df_off[df_off['国家地区'].isin(df_top5.index)]
df_off5sum=df_off5.groupby('国家地区')[['总额']].sum()
df_off5sum
```

```python
# 前5国家具销售总额
df_fur=df[df['商品门类']=='Furniture']
df_fur5=df_fur[df_fur['国家地区'].isin(df_top5.index)]
df_fur5sum=df_fur5.groupby('国家地区')[['总额']].sum()
df_fur5sum
```

```python
# 前5国科技销售总额
df_tec=df[df['商品门类']=='Technology']
df_tec5=df_tec[df_tec['国家地区'].isin(df_top5.index)]
df_tec5sum=df_tec5.groupby('国家地区')[['总额']].sum()
df_tec5sum
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# 画图
sns.set_style('whitegrid', {'font.sans-serif':['simhei', 'Arial']})
fig,(a,b,c)=plt.subplots(1,3,figsize=(15,4))
sns.barplot(x=df_off5sum.index,y='总额',data=df_off5sum,ax=a,palette='spring')
sns.barplot(x=df_fur5sum.index,y='总额',data=df_fur5sum,ax=b)
sns.barplot(x=df_tec5sum.index,y='总额',data=df_tec5sum,ax=c)
```

```python
df = df[['订单号', '国家地区', '商品门类', '总额']]
df['商品门类'].replace('Office Supplies', '办公', inplace=True)
df['商品门类'].replace('Furniture', '家具', inplace=True)
df['商品门类'].replace('Technology', '科技', inplace=True)

df_top5 = df.groupby('国家地区')[['总额']].sum().sort_values('总额', ascending=False).head(5)
df_top5 = df[df['国家地区'].isin(df_top5.index)]

fig,axs = plt.subplots(1, 3, figsize=(22, 7))
sns.barplot(x='国家地区', y='总额', data=df_top5[df_top5['商品门类']=='办公'], estimator=sum, ax=axs[0])
sns.barplot(x='国家地区', y='总额', data=df_top5[df_top5['商品门类']=='家具'], estimator=sum, ax=axs[1])
sns.barplot(x='国家地区', y='总额', data=df_top5[df_top5['商品门类']=='科技'], estimator=sum, ax=axs[2])
```

###     下载并读取气象观测数据，完成以下需求：

    1. 创建柱形图，展示平均风速排在前3的站点与平均风速；

    2. 使用内置蓝色调色板，透明度0.6，不显示误差线。 

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 20)
sh=pd.read_excel(r'C:\Users\YcAllenEffy\Desktop\气象观测.xlsx')
sh
```

```python
# 求前3站点
df=sh
df_top3 = df.groupby('站点')[['风速']].mean().sort_values('风速', ascending=False).head(3)
df_top3
```

```python
# 画图
sns.set_style('whitegrid', {'font.sans-serif':['simhei', 'Arial']})
sns.barplot(x=df_top3.index,y='风速',data=df_top3,palette='Blues',ci=None,alpha=0.6)
```

### 下载并读取气象观测数据，完成以下需求：

    1. 创建柱形图，展示总气压排名最后7名的站点与总气压值；

    2. 水平方向展示图形，坐标刻度值显示为普通数值，不显示误差线。 

```python
# 求末尾7站点,气压
df=sh
df_tail7 = df.groupby('站点')[['气压']].sum().sort_values('气压', ascending=False).tail(7)
df_tail7
```

```python
# 画图
sns.set_style('whitegrid', {'font.sans-serif':['simhei', 'Arial']})#初始化sns设置

plt.ticklabel_format(style='plain',axis='both')
sns.barplot(orient='h',y=df_tail7.index,x='气压',data=df_tail7,palette='Blues',ci=None,alpha=0.6)
```

### 下载并读取气象观测数据，完成以下需求：

    1. 创建柱形图，展示各月份的平均湿度；

    2. 透明度0.6，不显示误差线，调色板为summer；

    3. 设置x轴与y轴刻度的字体大小为xx-large，y轴刻度标签旋转90度。

```python
# 时间datetime格式化,列出月份列
df=sh
df['时间']=pd.to_datetime(df['时间'])
df['月份']=df['时间'].dt.month
df
```

```python
# 画图
sns.set_style('whitegrid', {'font.sans-serif':['simhei', 'Arial']})#初始化sns设置
plt.ticklabel_format(style='plain',axis='both')
g=sns.barplot(y='湿度',x='月份',data=df,palette='summer',ci=None,alpha=0.6)

# 避免冲突,强制完成画图,再设置xy轴标签样式
g.figure.canvas.draw()
g.set_yticks(g.get_yticks())

g.set_xticklabels(labels=g.get_xticklabels(),size='xx-large');
g.set_yticklabels(labels=g.get_yticklabels(),size='xx-large',rotation=90);
```

```python
df=sh


df['时间'] = pd.to_datetime(df['时间'])
df['月'] = df['时间'].dt.month

df_hum = df.groupby('月')[['湿度']].mean().sort_values('湿度', ascending=False)
g = sns.barplot(x=df_hum.index, y='湿度', data=df_hum, alpha=0.6, ci=None, palette='summer')
g.figure.canvas.draw()
g.set_xticks(g.get_xticks())
g.set_yticks(g.get_yticks())
g.set_xticklabels(labels=g.get_xticklabels(), size='xx-large')
g.set_yticklabels(labels=g.get_yticklabels(), rotation=90, size='xx-large');
```

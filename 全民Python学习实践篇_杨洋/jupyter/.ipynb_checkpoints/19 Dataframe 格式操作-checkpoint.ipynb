  describe 统计指标描述性信息df.describe()

  shape 行列信息 df.shape

  

  head 只显示头部几条 df.head(n)

  tail 只显示尾部几条df.head(n)

  

  显示所有行,列设置: pandas.set_option('display.max_rows',None)

  恢复所有行列设置:  pandas.reset_option('display.max_rows')
  
  设置不允许换行显示: pd.set_option('expand_frame_repr', False)
  
  设置小数的小数点后保留1位，显示数据集，重置精度。pandas.set_option('display.precision',1)
  

#### 删除行列

  删除行 df=df.drop('索引名') 默认axis=0,方法dorp 为新增df所以要赋值

  删除列 df=df.drop('列名',axis=1)

  按下标删除行df=df.drop(df.index[下标数])

  按下标删除列 df=df.drop(df.columns[下标数],axis=1)

  删除多行 df=df.drop(['索引1','索引2'])

  删除多列 df=df.drop(['列名1','列名2'],axis=1)

  掩码删除行 df=df.drop(df[df['列名']>100].index)



#### 增加行列

  增加列 df['新列名']='新值'

  插入位置新列 df.insert(插入位置,'新列名','数值')

  增加行 df.loc['新行名']=[各列值]

  df增加df行  append



#### 修改行列顺序

  修改列序 df=df[[新列序]] #列名必须和以前一样

              df=df.reindex(columns=[新列序],fill_value)

  修改行序 df.index=(新行序)

               df=df.reindex(index=[新行序],fill_value)

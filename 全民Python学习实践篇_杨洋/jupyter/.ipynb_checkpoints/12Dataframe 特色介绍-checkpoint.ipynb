### 一:灵活索引

  df['列一'] #取列一值

  df.loc['索引一'] #取索引一值

  自动按名对其

  运算时df1和df2自动按照表索引与列名对齐后运算

  df1+df2

  

#### 二:时间序列

  df['时间一':'时间二']  取时间一到时间二索引数据

  df['时间一'][df['列一']=='某'] 取列一等于某的时间一值

  df[df.index.hour==0] 取时间小时hour属性等于0的数据

#### 三:分组 合并方便

  df.groupby(df.index.date).agg(sum) 分组汇总

  df.rolling(3).mean() 3日均线

  pd.merge(df1,df2,left_on=['一','二'],right_on=['一','二'],how='left') 连接量表

#### 四:缺失值处理

  df[pd.isnull(df['一'])]

  df['一'].fillna('待察')

  

#### 五:文本处理

  df[df['一'].str.match(r'\w+仪')] 正则表达式

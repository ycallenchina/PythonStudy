### 计算相邻数据的时间差

    df['列名'].shift(1) 平移数据,向后平移数据一位.

    参数:fill_value,平移后的空值填充

    参数:axis=1 横向平移

  ##### 计算数据列的上下时间差

    df时间列的时间差,第一行时间差为第一行数值减去0

    df['时间列']-df['时间列'].shift(1,fill_value=0)  

    

  #### 时间差timedelta类型

   ##### 赋值timedelta

    pd.Timedelta(3,'D') 表示为时间差3天类型

      '3D5h'表示3天5小时

      W表示周,D表示天,h小时,T/m分,s秒,L毫秒,U微秒,N纳秒

   ##### 时间差的计算

    df['时间列']+pd.Timedelta('3D') 时间列统计加上3天时差

    

  #### 索引时间的计算

    df.shift(3,freq='D')

      D,W,h,s用法与时间差相同

### 读取 角色职业数据-9，利用网格搜索获取最优k和p，预测未知角色的职业。

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split ,GridSearchCV
from sklearn.metrics import accuracy_score

# 导入数据
file = r'C:\Users\YcAllenEffy\Desktop\fire_emblem_9.xlsx'
df_src = pd.read_excel(file, sheet_name=0)      # 已知数据
df_ans = pd.read_excel(file, sheet_name=1)  # 未知数据
```

```python
# 标准化已知数据

from sklearn.preprocessing import StandardScaler
mms=StandardScaler()
x=mms.fit_transform(df_src.iloc[:,2:])
y=df_src['职业']
```

```python
# 网格搜索和交叉验证

knn=KNeighborsClassifier()
gscv = GridSearchCV( knn, {'n_neighbors':range(1,10,2),'p':[1,2,float('inf')]}, cv=3 )
gscv.fit(x, y)
pd.DataFrame(gscv.cv_results_)
```

```python
gscv.best_params_
```

```python
# 获取最优参数值
k=gscv.best_params_['n_neighbors']
p1=gscv.best_params_['p']
```

```python
# 标准化未知数据
df_ans_S=mms.fit_transform(df_src.append(df_ans).iloc[:,2:])[-1:,:]
```

```python
# 预测未知数据
knn=KNeighborsClassifier(n_neighbors=k,p=p1)
model=knn.fit(x,y)
model.predict(df_ans_S)
```

```python
# 标准答案
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV

# 导入数据
file = r'C:\Users\YcAllenEffy\Desktop\fire_emblem_9.xlsx'
df_src = pd.read_excel(file, sheet_name=0)      # 已知数据
df_unknown = pd.read_excel(file, sheet_name=1)  # 未知数据

# 已知数据标准化
ss = StandardScaler()
x = ss.fit_transform( df_src.iloc[ :, 2:12 ] )
y = df_src['职业']

# 网格搜索和交叉验证
knn = KNeighborsClassifier()
gscv = GridSearchCV( knn, {'n_neighbors':range(1,10,2), 'p':[1,2,float('inf')]}, cv=3 )
gscv.fit(x, y)
pd.DataFrame(gscv.cv_results_)
k = gscv.best_params_['n_neighbors']
p = gscv.best_params_['p']

# 未知数据标准化
x_unknown = ss.transform(df_unknown.iloc[:, 2:12])
# 预测未知职业
knn = KNeighborsClassifier(n_neighbors=k, p=p)
model = knn.fit(x, y)
res = model.predict( x_unknown )[0]
print('预测的职业是：', res)
```

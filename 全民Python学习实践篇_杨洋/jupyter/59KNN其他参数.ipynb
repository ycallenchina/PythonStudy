### 读取 iris鸢尾花数据，将数据集拆分为训练集和测试集，设置KNN距离权重，通过网格搜索查找最优k和p值，通过测试集验证准确率。

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 20)
sh=pd.read_csv(r'C:\Users\YcAllenEffy\Desktop\iris.csv')
df=sh.copy()
df
```

```python
# 导入包
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV , train_test_split
from sklearn.metrics import accuracy_score
```

```python
# 拆分数据集
df_train,df_test=train_test_split(df)
df_train
```

```python
# 网格搜索
knn=KNeighborsClassifier(weights='distance')
gscv=GridSearchCV(knn, {'n_neighbors':range(1,10,2),'p':[1,2,float('inf')]}, cv=3 )

x=df_train.iloc[:,:-1]
y=df_train['种类']
gscv.fit(x, y)
pd.DataFrame(gscv.cv_results_)
```

```python
# 取网格搜索最优值
k=gscv.best_params_['n_neighbors']
p1=gscv.best_params_['p']
```

```python
# 预测未知数据
knn=KNeighborsClassifier(n_neighbors=k,p=p1,weights='distance')
model=knn.fit(x,y)
ans=model.predict(df_test.iloc[:,:-1])
```

```python
# 准确率
accuracy_score(df_test['种类'],ans)
```

```python
# 标准答案
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score


# 导入数据
file = r'C:\Users\YcAllenEffy\Desktop\iris.csv'
df_src = pd.read_csv(file)

# 拆分训练集与测试集
df_train, df_test = train_test_split(df_src)

# 标准化
ss = StandardScaler()
x_train = ss.fit_transform( df_train.iloc[ :, :4 ] )
y_train = df_train['种类']
x_test = ss.transform( df_test.iloc[ :, :4 ] )

# 网格搜索交叉验证
knn = KNeighborsClassifier(weights='distance')
gscv = GridSearchCV( knn, {'n_neighbors':range(1,10,2), 'p':[1,2,float('inf')]}, cv=3 )
gscv.fit(x_train, y_train)
pd.DataFrame(gscv.cv_results_)
k = gscv.best_params_['n_neighbors']
p = gscv.best_params_['p']

# 测试集验证
knn = KNeighborsClassifier(n_neighbors=k, p=p, weights='distance')
model = knn.fit(x_train, y_train)
y_predict = model.predict( x_test )
acc = accuracy_score( df_test['种类'], y_predict )
print( f'测试集预测准确率是 {acc}' )
```

```python
# 标准答案
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
import sklearn.metrics as metrics

# 导入数据
file = r'C:\Users\YcAllenEffy\Desktop\iris.csv'
df_src = pd.read_csv(file)

# 拆分训练集与测试集
df_train, df_test = train_test_split(df_src)

# 标准化
ss = StandardScaler()
x_train = ss.fit_transform( df_train.iloc[ :, :4 ] )
y_train = df_train['种类']
x_test = ss.transform( df_test.iloc[ :, :4 ] )

# 网格搜索交叉验证
knn = KNeighborsClassifier(weights='distance')
gscv = GridSearchCV( knn, {'n_neighbors':range(1,10,2), 'p':[1,2,float('inf')]}, cv=3 )
gscv.fit(x_train, y_train)
pd.DataFrame(gscv.cv_results_)
k = gscv.best_params_['n_neighbors']
p = gscv.best_params_['p']

# 测试集验证
knn = KNeighborsClassifier(n_neighbors=k, p=p, weights='distance')
model = knn.fit(x_train, y_train)
y_predict = model.predict( x_test )
acc = accuracy_score( df_test['种类'], y_predict )
print( f'测试集预测准确率是 {acc}' )

# 求分值
print("召回率", metrics.recall_score(df_test['种类'], y_predict, average='weighted'))
print("精确率", metrics.precision_score(df_test['种类'], y_predict, average='weighted'))
print("F1分值", metrics.f1_score(df_test['种类'], y_predict, average='weighted'))

# 未知数据
# 6.3, 2.8, 5.1, 1.5 ?
df_unknown = pd.DataFrame({ '萼片长': [6.3],'萼片宽':[2.8], '花瓣长':[5.1], '花瓣宽':[1.5], '种类':'?' })
# 未知数据标准化
x_unknown = ss.transform(df_unknown.iloc[:, :4])
# 未知数据预测
res = model.predict( x_unknown )[0]
print( f'未知数据的种类是 {res}' )
```

提示：urlopen返回的响应对象调用getcode方法可以获取到服务器的响应码，如果是200代表响应正常。

```python
from urllib.request import urlopen
from functools import wraps
def print_response(f):
    @wraps(f)
    def add_print(url):
        a=f(url)
        print(a.getcode())
        print(a.read().decode('utf-8'))
        return a
    return add_print

@print_response
def get_response(url):
    response = urlopen(url=url)
    
    return response


        
url = 'http://myip.ipip.net'
# get_response=print_response(get_response)
get_response(url)

# response = get_response(url)
# print(response.getcode())
# print(response.read().decode('utf-8'))
```

```python
# 标准答案

from urllib.request import urlopen

from functools import wraps



def get_response_state(func):

    @wraps(func)

    def inner(url):

        response = func(url)

        print('请求地址：', url)

        print('响应状态：', response.getcode())

        return response

    return inner



@get_response_state

def get_response(url):

    response = urlopen(url=url)

    return response





url = 'http://myip.ipip.net'

response = get_response(url)

print(response.read().decode('utf-8'))
```

```python
# 标准答案

import random

import time



array = [random.randint(0, 10000) for _ in range(10000)]



def bubble_sort(array):

    n = len(array)

    for i in range(n - 1):

        for j in range(0, n - i - 1):

            if array[j] > array[j + 1]:

                array[j], array[j + 1] = array[j + 1], array[j]

    return array



def time_func(func):

    def inner(array):

        start = time.time()

        func(array)

        end = time.time()

        return end - start

    return inner



func1 = time_func(bubble_sort)

func2 = time_func(sorted)

print('内置排序执行时间：', func2(array))

print('冒泡排序执行时间：', func1(array))
```

```python
#标准答案
import random
import time
from functools import wraps


array = [random.randint(0, 10000) for _ in range(10000)]

def time_func(func):
    @wraps(func)
    def inner(array):
        start = time.time()
        func(array)
        end = time.time()
        return end - start
    return inner


@time_func
def bubble_sort(array):
    n = len(array)
    for i in range(n - 1):
        for j in range(0, n - i - 1):
            if array[j] > array[j + 1]:
                array[j], array[j + 1] = array[j + 1], array[j]
    return array


@time_func
def my_sorted(array):
    sorted(array)


print('内置排序执行时间：', my_sorted(array))
print('冒泡排序执行时间：', bubble_sort(array))
```

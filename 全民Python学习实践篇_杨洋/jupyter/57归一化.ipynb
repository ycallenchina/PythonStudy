```python
# 标准答案
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import MinMaxScaler

# 导入数据
file =r'C:\Users\YcAllenEffy\Desktop\fire_emblem_7.xlsx'
df_src = pd.read_excel(file, sheet_name=0)      # 已知数据
df_unknown = pd.read_excel(file, sheet_name=1)  # 未知数据
# 数据规范化
mms = MinMaxScaler()
x = mms.fit_transform( df_src.iloc[:,2:12] )
y = df_src['职业']
# 交叉验证得到k
knn = KNeighborsClassifier()
gscv = GridSearchCV( knn, {'n_neighbors':range(1,10,2)}, cv=3, 
                    return_train_score=True )
gscv.fit(x, y)
pd.DataFrame(gscv.cv_results_)
k = gscv.best_params_['n_neighbors']
# 重新训练模型，预测未知角色
knn = KNeighborsClassifier( n_neighbors=k )
model = knn.fit(x, y)
x_unknown = mms.fit_transform(df_src.append(df_unknown).iloc[:, 2:12] )[-1:, :]
res = model.predict( x_unknown )[0]
print('未知角色的职业是：', res,'k值为',k)
```

```python
# 标准答案
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV

# 导入数据
file = r'C:\Users\YcAllenEffy\Desktop\fire_emblem_6.xlsx'
df_src = pd.read_excel(file, sheet_name=0)      # 已知数据
df_unknown = pd.read_excel(file, sheet_name=1)  # 未知数据

# 拆分训练接与测试集
df_train, df_test = train_test_split(df_src)

# 对训练集进行网格搜索
knn = KNeighborsClassifier()
gscv = GridSearchCV( knn, {'n_neighbors':range(1,5)}, cv=5 )
x = df_train.iloc[ :, 2:12 ]
y = df_train['职业']
gscv.fit(x, y)
# 获取最优参数值
pd.DataFrame(gscv.cv_results_)
k = gscv.best_params_['n_neighbors']

# 使用测试集验证
knn.n_neighbors = k
model = knn.fit(x, y)
y_predict = model.predict( df_test.iloc[:, 2:12] )
acc = accuracy_score( df_test['职业'], y_predict )
print( f'当K值是{k}时，测试集预测准确率是 {acc}' )

# 对未知数据分类
res = model.predict( df_unknown.iloc[ :, 2:12 ] )[0]
print('尤利乌斯的职业是：', res)
```

```python
# 标准答案
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV

# 导入数据
file = r'C:\Users\YcAllenEffy\Desktop\wine_quality.xlsx'
df_src = pd.read_excel(file, sheet_name=0)      # 已知数据
df_unknown = pd.read_excel(file, sheet_name=1)  # 未知数据

# 拆分训练接与测试集
df_train, df_test = train_test_split(df_src)

# 对训练集进行网格搜索
knn = KNeighborsClassifier()
gscv = GridSearchCV( knn, {'n_neighbors':range(1,10,2)}, cv=10 )
x = df_train.iloc[ :, 0:6 ]
y = df_train['品质']
gscv.fit(x, y)
# 获取最优参数值
pd.DataFrame(gscv.cv_results_)
k = gscv.best_params_['n_neighbors']

# 使用测试集验证
knn.n_neighbors = k
model = knn.fit(x, y)
y_predict = model.predict( df_test.iloc[:, 0:6] )
acc = accuracy_score( df_test['品质'], y_predict )
print( f'当K值是{k}时，测试集预测准确率是 {acc}' )

# 对未知数据预测
res = model.predict( df_unknown.iloc[ :, 0:6 ] )[0]
print('未知葡萄酒的品质是：', res)
```

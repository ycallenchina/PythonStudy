```python
### 继续使用练习1, 预测当运费是30的总额值。
```

```python
#解法一:用sklearn模块
from sklearn.linear_model import LinearRegression

rm=LinearRegression()
rm.fit(df_c[['运费']],df_c[['总额']])
b=rm.predict([[30,]])
b
```

```python
# 解法二:用sklearn模块

import scipy.stats as st
q=st.linregress(df_c['运费'],df_c['总额'])
斜率=q.slope
截距=q.intercept

预测值=lambda x,斜率,截距 : 斜率*x+截距

预测值(30,斜率,截距)


```

```python
# 标准答案
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid', {'font.sans-serif':['simhei', 'Arial']})
plt.rcParams['axes.unicode_minus'] = False

df = sh.copy
dfc = df[df['商品子类']=='Art']
dfc = dfc[['总额', '运费', '折扣', '营利', '数量']]

import scipy.stats as st
a, b, r2, p, std = st.linregress( dfc['运费'], dfc['总额'] )
print(a * 30 + b)
```

```python
###  继续使用练习2的数据集，使用 LinearRegression 模型拟合运费与总额，预测总体的估计总额，查看准确率。
```

```python
from sklearn.linear_model import LinearRegression

rm=LinearRegression()
rm.fit(df_c[['运费']],df_c[['总额']])

df_c['总额估计']=rm.predict(df_c[['运费']])
df_c
```

```python
rm.score(df_c[['运费']],df_c[['总额']])
```

```python
# 标准答案
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid', {'font.sans-serif':['simhei', 'Arial']})
plt.rcParams['axes.unicode_minus'] = False

df = sh.copy()
dfc = df[df['商品子类']=='Art']
dfc = dfc[['总额', '运费', '折扣', '营利', '数量']]

from sklearn.linear_model import LinearRegression
rm = LinearRegression()
rm.fit( dfc[['运费']], dfc[['总额']] )
df['估计总额'] = rm.predict( df[['运费']] )
rm.score( dfc[['运费']], dfc[['总额']] )
```

```python
# 标准答案
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 导入数据
file = r'C:\Users\YcAllenEffy\Desktop\fire_emblem_4.xlsx'
df_src = pd.read_excel(file, sheet_name=0)      # 已知数据
df_unknown = pd.read_excel(file, sheet_name=1)  # 未知数据

# 将已知数据拆分为训练集和测试集
df_train, df_test = train_test_split( df_src, test_size=0.2, random_state=42 )

# 用训练集训练模型，用测试集测试并得到最高准确率的K值
x = df_train.iloc[:, 2:12]
y = df_train['职业']

for k in range(1, 20, 2):
    knn = KNeighborsClassifier(n_neighbors = k)
    model = knn.fit(x, y)
    # 对测试集进行预测，求准确率
    y_predict = model.predict( df_test.iloc[:, 2:12] )
    acc = accuracy_score( df_test['职业'], y_predict )
    print( f'当K值是{k}时，预测准确率是 {acc}' )

# 设置最高准确率的K值，重新训练，预测未知数据
knn.n_neighbors = 1
model = knn.fit(x, y)
res = model.predict( df_unknown.iloc[ :, 2:12 ] )[0]
print('辛格尔德的职业是：', res)
```

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 导入数据
file = r'C:\Users\YcAllenEffy\Desktop\fire_emblem_5.xlsx'
df_src = pd.read_excel(file, sheet_name=0)      # 已知数据
df_unknown = pd.read_excel(file, sheet_name=1)  # 未知数据

# 将已知数据拆分为训练集和测试集
df_train, df_test = train_test_split( df_src, test_size=0.2, random_state=66 )

# 用训练集训练模型，用测试集测试并得到最高准确率的K值
x = df_train.iloc[:, 2:11]
y = df_train['速度']

for k in range(1, 10, 2):
    knn = KNeighborsClassifier(n_neighbors = k)
    model = knn.fit(x, y)
    # 对测试集进行预测，求准确率
    y_predict = model.predict( df_test.iloc[:, 2:11] )
    print(y_predict)
    acc = accuracy_score( df_test['速度'], y_predict )
    print( f'当K值是{k}时，预测准确率是 {acc}' )

knn.n_neighbors = 1
model = knn.fit(x, y)
res = model.predict( df_unknown.iloc[ :, 2:11 ] )[0]
print('瓦尔基里的速度是：', res)
```

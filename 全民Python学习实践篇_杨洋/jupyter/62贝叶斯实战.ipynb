### 读取 外卖评论数据 ，使用朴素贝叶斯分类器预测下面的评论是好评还是差评：

    "直接整无语了 0.2带餐具，就一双筷子，连勺子都不愿意给，整个实惠套餐就几个鱼蛋，剩下大多是豆芽"

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 20)
sh=pd.read_csv(r'C:\Users\YcAllenEffy\Desktop\外卖评论.csv')
df=sh.copy()
df
```

```python
#删除空文本记录
df.dropna(subset=['评论'],inplace=True)
```

```python
#分词
import jieba
df['分词']=df['评论'].map(lambda s: " ".join(jieba.cut(s)))
df
```

```python
#停止词表
stopwords = []
with open (r'C:\Users\YcAllenEffy\Desktop\baidu-stopword.txt', encoding='utf-8') as f:
   stopwords = f.readlines()
stopwords = [ s.strip() for s in stopwords ] #去两端空格,去换行符
```

```python
#向量化词频
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer

#划分测试训练
df_train,df_test =train_test_split(df,random_state=42)
#向量化
Vect=CountVectorizer(stop_words=stopwords)
x_train = Vect.fit_transform(df_train['分词'])
```

```python
#查看向量化结果
Vect.vocabulary_
# print(x_train)
```

```python
#贝叶斯训练
from sklearn.naive_bayes import MultinomialNB
bys=MultinomialNB(alpha=1)
model=bys.fit(x_train,df_train['分类'])
```

```python
#测试
x_test=Vect.transform(df_test['分词'])
y_test=model.predict(x_test)

from sklearn import metrics
metrics.accuracy_score(y_test,df_test['分类'])
```

```python
# 标准答案
#读取数据
import pandas as pd
df = sh

# 分词
import jieba
df['分词'] = df['评论'].map( lambda s: ' '.join( jieba.cut(s) ) )
# 读取停止词
# stopwords = []
# with open (r'.\baidu-stopword.txt', encoding='utf-8') as f:
#    stopwords = f.readlines()
# stopwords = [ s.strip() for s in stopwords ]

# 拆分测试集与训练集
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split( df, random_state=42 )

# 训练集向量化
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer( stop_words = stopwords )
x_train = vect.fit_transform( df_train['分词'] )

# 创建贝叶斯分类器，训练模型
from sklearn.naive_bayes import MultinomialNB
bys = MultinomialNB()
model = bys.fit( x_train, df_train['分类'] )

# 对测试集预测，求准确性
x_test = vect.transform( df_test['分词'] )
y_test = model.predict( x_test )
from sklearn import metrics
print('准确性：', metrics.accuracy_score(y_test, df_test['分类']) )

# 预测未知数据
s = "直接整无语了 0.2带餐具，就一双筷子，连勺子都不愿意给，整个实惠套餐就几个鱼蛋，剩下大多是豆芽"
x_unknown = " ".join(jieba.cut(s))
x_predict = vect.transform( [x_unknown] )
res = model.predict( x_predict )
print('未知评论的类型是：', res[0])
```
